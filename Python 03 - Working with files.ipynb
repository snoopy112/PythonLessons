### Работа с файлами

**Работа с текстовыми файлами**

Чтение содержимого файла происходит с помощью конструкции **open('имя_файла', 'r')**, где ключ **'r'** обозначает чтение файла.

```python
file = open('example.txt', 'r') # открытие файла для чтения
```

```python
data = file.read() # чтение данных из файла
data
```

```python
file.close() # закрытие файла
```

Запись данных в файл происходит с помощью конструкции **open('имя_файла', 'w')**, где ключ **'w'** обоначает запись в файл. При этом содержимое файла стирается полностью перед записью.

```python
data = data.upper() # заменям символы на заглавные
data
```

```python
file = open('example1.txt', 'w')
```

```python
file.write(data) # запись данных из переменной data в файл
```

```python
file.close() # закрытие файла
```

Ключи могут быть:<br>
**'a'** - при записи позволяет не перезаписывать файл, а добавить строки в конец<br>
**'a+'** - файл открывается для записи и чтения, позволяя как добавлять в конец файла, так и читать содержимое<br>
**'r'** - чтение текстового файла<br>
**'rb'** - чтение бинарного файла (read-binary)<br>
**'r+'** - считывает файл и стирает содерживое, позволяя перезаписывать файл<br>
**'w'** - запись текстовой переменной в файл<br>
**'wb'** - запись бинарной переменной в файл (write-binary)<br>
**'w+'** - то же, что и "w+r"

**Работа с файлом с помощью контекстного менеджера (использование оператора with)**

В Python имеется встроенный инструмент, применяя который вы можете заметно упростить чтение и редактирование файлов. Оператор **with** создает диспетчер контекста, который автоматически закрывает файл по окончанию работы в нем. 

```python
with open('example.txt', 'r') as file:
    for line in file:
        print(line)
```

**Работа с изображениями с помощью библиотеки matplotlib**
